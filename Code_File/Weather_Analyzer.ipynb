```python
df5['dt'] = pd.to_datetime(df5['dt'])
df5=df5.set_index('dt')
annual_df5= df5.resample('A').mean()

```

```python
annual_df5=annual_df5.reset_index()
```

```python
annual_df5['dt']=annual_df5['dt'].dt.year
```

```python
annual_df5.head()
```

#### Let the baseline time for the temperature be between 1951 and 1980

```python
#datetime
annual_df5['dt'] = pd.to_datetime(annual_df5['dt'])
# Index Setting
annual_df5 = annual_df5.set_index('dt')
baseline_data = annual_df5['1951':'1980']
baseline_temp = baseline_data['LandAverageTemperature'].mean()
print(f"The baseline temperature is: {baseline_temp:.2f} degree Celsius")
```

```python
# Calculating deviation
annual_df5['anomaly'] = annual_df5['LandAverageTemperature'] - baseline_temp
print(annual_df5[['LandAverageTemperature', 'anomaly']].tail())
```

```python
print(annual_df5[['LandAverageTemperature', 'anomaly']].head())
```

```python
annual_df5['rolling_anomaly'] = annual_df5['anomaly'].rolling(window=10).mean()
print(annual_df5['rolling_anomaly'])
```
